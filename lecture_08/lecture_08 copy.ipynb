We can cut down on redundant code by using a `for` loop.

````python
good_movies = [] # create empty list

cut_off = 7.0 # minimum rating for "good movie"

movie_titles = ["The Umbrellas of Cherbourg", "M3gan", "Psycho"]
movie_ratings = [7.8, 6.3, 8.5]

# how should we proceed?

index = 0

for rating in movie_ratings:

    if rating >= cut_off:
        good_movies = good_movies + [movie_titles[index]]
    
    index = index + 1

print(good_movies)
````

**Example**: Controlling accuracy. Consider the function

$$f(x) = \frac{1}{x}$$

- If $x$ is large enough, then $f(x)$ becomes very small

- We can continue to run values until $0 < f(x) \leq \epsilon$
- Here the value $\epsilon$ is a **threshold**

Similar logic for any algorithm that continues executing until an accuracy criteria is satisfied

````python
# Specify a threshold level

eps = 0.025

x_list = []
y_list = []


x = 1
while 1/x > eps:
    x_list.append(x)
    y_list.append(1/x)
    x = x + 1

plt.scatter(x = x_list, y = y_list)
plt.title("Graph of f(x) = 1/x")
plt.xlabel('x')
plt.ylabel("y")
plt.show()
````

````python
# The next example simulates a discrete-time Bernoulli process.
# For each time step, we will perform an experiment leading either 
# to "Success" or "Failure"

import random 

possible_outcomes = ["Success", "Failure"]

# experiment
outcome = random.choice(possible_outcomes)
print(outcome)
````

````python
# While Loop

success_count = 0
t = 0.0
final_time = 1.0
dt = 0.1

# while t != final_time:  # <--- WARNING: using this instead will cause infinite loop!

while abs(t - final_time) > 1e-13:
    # Experiment
    random_outcome = random.choice(possible_outcomes)

    # If it was a success, count it by increasing the success_count variable
    if random_outcome == "Success":
        success_count = success_count + 1

    # Increase time to the next value
    t = t + dt

print(success_count)

    
````

## <span style="color:red; font-size:25px"> IV. Interrupting `for` loops </span>

<font size = "4">

Often, instead of a `while` loop, we can use a `for` loop with a `break` statement.

````python
# Specify a threshold level

eps = 0.025

x_list = []
y_list = []


x = 1

max_points = 100000

for k in range(max_points):
    if 1/x <= eps:
        break

    x_list.append(x)
    y_list.append(1/x)
    x = x + 1

plt.scatter(x= x_list, y = y_list)
plt.title("Graph of f(x) = 1/x")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
````

<font size = "4">

Example: stop a `for` loop when an invalid type is encountered

````python
list_mixed = [1, 2, None, 4, "missing", 11, 2.2]

list_sum = 0
for value in list_mixed:
    if (type(value) != int) and (type(value) != float):
        print("Stopping sum computation: Non-numerical element in list!")
        break

    list_sum = list_sum + value

print(list_sum)
````

<font size = "4">

In other situations, you might want to silently skip over an invalid input without terminating the entire loop

````python
list_mixed = [1, 2, None, 4, "missing", 11, 2.2, None, "missing", 3]

list_sum = 0
for value in list_mixed:
    if (type(value) != int) and (type(value) != float):
        print("Warning: Non-numerical element in list!")
        continue

    list_sum = list_sum + value

print(list_sum)
````
